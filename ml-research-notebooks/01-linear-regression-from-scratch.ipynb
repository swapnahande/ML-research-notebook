# ðŸ“˜ Linear Regression from Scratch
In this notebook, we will:
- Generate synthetic data
- Implement linear regression using Mean Squared Error (MSE)
- Train using Gradient Descent
- Visualize the loss and regression line

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating the data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.title('Generated Data')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Add bias term
X_b = np.c_[np.ones((100, 1)), X]

# Initialize weights
theta = np.random.randn(2,1)
learning_rate = 0.1
n_iterations = 1000
m = 100

# Gradient Descent
losses = []
for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients
    loss = np.mean((X_b.dot(theta) - y)**2)
    losses.append(loss)
```

```python
# Plot loss
plt.plot(losses)
plt.title('Loss over iterations')
plt.xlabel('Iteration')
plt.ylabel('MSE Loss')
plt.grid(True)
plt.show()
```

```python
# Visualize final line
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta), color='red')
plt.title('Linear Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
